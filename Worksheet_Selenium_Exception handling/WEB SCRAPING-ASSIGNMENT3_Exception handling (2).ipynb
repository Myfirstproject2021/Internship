```python
product_title=[]
product_details=[]
search_smartphone=driver.find_elements_by_xpath('//div[@class="_4rR01T"]')
for i in search_smartphone:
    product_title.append(i.text.split('('))
    
search_smartphone_details=driver.find_elements_by_xpath('//ul[@class="_1xgFaf"]')
for i in search_smartphone_details:   
    try:
        product_details.append(i.text.split('\n'))           
    except:
         product_details.append('-')    
```

```python
df2=pd.DataFrame(product_title)
#len(df2)
len(df)
```

```python
import csv
# name of csv file  
filename = "SmartPhone_Details.csv"
fields=['Product Name','Colour','Storage','Display','Camara','Battery','Processor','Warrantly']
rows=Product
# writing to csv file  
with open(filename, 'w') as csvfile:  
    # creating a csv writer object  
    csvwriter = csv.writer(csvfile)  
          
    # writing the fields  
    csvwriter.writerow(fields)  
          
    # writing the data rows  
    csvwriter.writerows(rows)
```
