```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

QUESTION 2_IMDB_ALLMOVIES

```python
page=requests.get('https://www.imdb.com/list/ls055592025/')
soup=BeautifulSoup(page.content)
Titles=[]
for i in soup.find_all('h3', class_="lister-item-header"):
    Titles.append(i.text.split('\n'))
Rating=[]
for i in soup.find_all('div', class_="ipl-rating-star small"):
    Rating.append(i.text.split('\n'))
df=pd.DataFrame(Titles)
df2=pd.DataFrame(Rating)
frames = [df, df2]
ANS_2_Table = pd.concat(frames,join='inner',axis=1)
ANS_2_Table.columns =['Empty', 'S.No', 'Title', 'Year','Empty','Empty','Empty','Empty','Empty','Empty','Empty','Empty','Empty','Rating','Empty',]
ANS_2_Table.drop('Empty', inplace=True, axis=1)
#pd.set_option('display.max_rows', None)
ANS_2_Table
```

QUESTION 3_IMDB_INDIAN_MOVIES

```python
page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=461131e5-5af0-4e50-bee2-223fad1e00ca&pf_rd_r=QDTJYFRGB7JVRX6Y0X62&pf_rd_s=center-1&pf_rd_t=60601&pf_rd_i=india.toprated&ref_=fea_india_ss_toprated_india_tr_india250_sm')
soup_indiamovies=BeautifulSoup(page.content)
INDIAN_Titles=[]
for i in soup_indiamovies.find_all('td', class_="titleColumn"):
    INDIAN_Titles.append(i.text.split('\n'))
df=pd.DataFrame(INDIAN_Titles)
INDIAN_RATING=[]
for i in soup_indiamovies.find_all('td', class_="ratingColumn imdbRating"):
    INDIAN_RATING.append(i.text.split('\n'))
df2=pd.DataFrame(INDIAN_RATING)
ANS_3_IndianMovies=pd.concat([df,df2],axis=1)
ANS_3_IndianMovies.columns=['DROP','RANK','NAME','YEAR','DROP','DROP','RATING','DROP']
ANS_3_IndianMovies.drop('DROP', inplace=True, axis=1) 
ANS_3_IndianMovies = ANS_3_IndianMovies.iloc[:100,:]
ANS_3_IndianMovies
```

QUESTION 4_MEESHO_PROODUCTS

```python
page=requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
soup_Meesho=BeautifulSoup(page.content)
Product=[]
for i in soup_Meesho.find_all('p', class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    Product.append(i.text.split(','))
Product_Prize=[]
for i in soup_Meesho.find_all('h5', class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    Product_Prize.append(i.text.split('₹'))
Product_Discount=[]
for i in soup_Meesho.find_all('span', class_="Text__StyledText-sc-oo0kvp-0 lnonyH"):
    Product_Discount.append(i.text)
df=pd.DataFrame(Product)
df2=pd.DataFrame(Product_Prize)
df3=pd.DataFrame(Product_Discount)
frames = [df, df2, df3]

ANS_4_Table = pd.concat(frames,join='inner',axis=1)
ANS_4_Table.columns =['Product','0','Orginal Prize', 'Discounted Prize']
ANS_4_Table.drop('0', inplace=True, axis=1)
#pd.set_option('display.max_rows', None)
ANS_4_Table
```

QUESTION_7_scrape details of all the posts from coreyms.com.

```python
page=requests.get('https://coreyms.com/')
soup_COREYMS=BeautifulSoup(page.content)
HEADER=[]
for i in soup_COREYMS.find_all('h2', class_="entry-title"):
    HEADER.append(i.text)
DATE=[]
for i in soup_COREYMS.find_all('time', class_="entry-time"):
    DATE.append(i.text)
CONTENT=[]
for i in soup_COREYMS.find_all('div', class_="entry-content"):
   CONTENT.append(i.text.split('\n'))
VEDIO=[]
for i in soup_COREYMS.find_all('iframe', class_="youtube-player"):
   VEDIO.append(i['src'])
df1=pd.DataFrame(HEADER)
df2=pd.DataFrame(DATE)
df3=pd.DataFrame(CONTENT)
df4=pd.DataFrame(VEDIO)

df = pd.concat([df1, df2], axis=1)
df= pd.concat([df, df3], axis=1)
df=pd.concat([df, df4], axis=1)

cols = [ 4, 5, 6, 7, 8]
df.drop(df.columns[cols], axis=1,inplace=True)
df.reset_index(drop=True)
df.columns =['HEADER', 'DATE','0','CONTENT','0','VEDIO']
df.drop('0', inplace=True, axis=1)
df
```

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
Soup_dineout=BeautifulSoup(page.content)
RESTURENT_NAME=[]
for i in Soup_dineout.find_all('a',class_="restnt-name ellipsis"):
   RESTURENT_NAME.append(i.text)
RESTURENT_LOCATION=[]
for i in Soup_dineout.find_all('div',class_="restnt-loc ellipsis"):
   RESTURENT_LOCATION.append(i.text)
RESTURENT_CUSINE=[]
for i in Soup_dineout.find_all('div',class_="detail-info"):
   RESTURENT_CUSINE.append(i.text.split(" | "))
RESTURENT_RATING=[]
for i in Soup_dineout.find_all('div',class_="restnt-rating rating-4"):
   RESTURENT_RATING.append(i.text)
RESTURENT_IMAGE=[]
for i in Soup_dineout.find_all('img',class_="no-img"):
   RESTURENT_IMAGE.append(i['data-src'])
df=pd.DataFrame(RESTURENT_NAME)
df1=pd.DataFrame(RESTURENT_LOCATION)
df2=pd.DataFrame(RESTURENT_CUSINE)
df3=pd.DataFrame(RESTURENT_RATING)
df4=pd.DataFrame(RESTURENT_IMAGE)
frames = [df, df1, df2, df3, df4 ]
ANS_9_DINEOUT = pd.concat(frames, axis=1)
ANS_9_DINEOUT.columns=['NAME', 'LOCATION','0','CUSINE','RATING','IMG']
ANS_9_DINEOUT.drop('0', inplace=True, axis=1)
ANS_9_DINEOUT
```
