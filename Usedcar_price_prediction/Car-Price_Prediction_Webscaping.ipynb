```python
def find_duplicates(sequence):
  first_seen = set()
  first_seen_add = first_seen.add  
  duplicates = set(i for i in sequence if i in first_seen or first_seen_add(i) )
  # turn the set into a list (as requested)
  return duplicates 


print(find_duplicates(product_urls_2))

#This shows that There are no duplicates
```

```python
import csv
# name of csv file  
filename = "OLX_product_urls_2.csv"
fields=['1000_url']
rows=product_urls_2
# writing to csv file  
with open(filename, 'w') as csvfile:  
    # creating a csv writer object  
    csvwriter = csv.writer(csvfile)  
          
    # writing the fields  
    csvwriter.writerow(fields)  
          
    # writing the data rows  
    csvwriter.writerows(rows)
```

```python
#Make empty lists
Brand = []   
delear_autonity = []
Fuel = []
KMS = []
TYPE = []
No_Owners = []
Location = []
Engine_CAPACITY = [] 
Colour = []
insurence=[]
Aux=[]
Tyre=[]
Power_windows=[]
airbags=[]
Power_Stering=[]
Lock=[]
prize=[]

for url in product_urls_2:
    driver.get(url)
    time.sleep(2)
    
    # Extracting Brand from xpath
    try:
        brand = driver.find_element_by_xpath('//div[@class="_35xN1"]')    
        Brand.append(brand.text)
    except:
        Brand.append('-')    
    time.sleep(1)
    
    
    try:
        name = driver.find_element_by_xpath('//span[@class="_2LMSU _1eesT"]')      
        delear_autonity.append(name.text)
    except:
        delear_autonity.append('-')
        
    
    time.sleep(1)    
    # Extracting Ratings from xpath    
    try:
        fuel = driver.find_element_by_xpath('//div[@class="_3qDp0"]')  
        Fuel.append(fuel.text)
    except:
        Fuel.append('-')
    time.sleep(1)
    # Extracting No of Ratings from xpath
    try:
        kms = driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/div/div[5]/div[1]/div[1]/div[1]/div/div[4]/div[2]/div')  
        KMS.append(kms.text)
    except:
        KMS.append('-')
    
    time.sleep(1)
    # Extracting price from xpath    
    try:
        Type = driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/div/div[5]/div[1]/div[1]/div[1]/div/div[4]/div[3]/div')  
        TYPE.append(Type.text)
    except:
        TYPE.append('-')
    
    try:
        noofowners= driver.find_element_by_xpath("//div[contains(text(),'Owner')]/following::div")  
        No_Owners.append(noofowners.text)
    except:
        No_Owners.append('-') 

    try:
        location= driver.find_element_by_xpath("//div[contains(text(),'Location')]/following::div")  
        Location.append(location.text)
    except:
        Location.append('-') 
        
    try:
        engine= driver.find_element_by_xpath(("//div[contains(text(),'Engine Capacity')]"))  
        Engine_CAPACITY.append(engine.text)
    except :
        Engine_CAPACITY.append('-') 
    
    try:
        COLOUR= driver.find_element_by_xpath("//div[contains(text(),'Color')]")
        Colour.append(COLOUR.text)
    except:
        Colour.append('-') 
    
    try:
        INSURENCE= driver.find_element_by_xpath("//div[contains(text(),'Insurance')]")  
        insurence.append(INSURENCE.text)
    except :
        insurence.append('-') 
    
    try:
        TYRE= driver.find_element_by_xpath("//div[contains(text(),'Tyre Condition')]") 
        Tyre.append(TYRE.text)
    except :
        Tyre.append('-') 
        
    try:
        AUX= driver.find_element_by_xpath("//div[contains(text(),'Aux Compatibility')]")
        Aux.append(AUX.text)
    except :
        Aux.append('-') 
        
    try:
        POWERWINDOWS= driver.find_element_by_xpath("//div[contains(text(),'Power Windows')]")  
        Power_windows.append(POWERWINDOWS.text)
    except:
        Power_windows.append('-') 
    
    try:
        AIRBAGS= driver.find_element_by_xpath("//div[contains(text(),'Number of Airbags')]")
        airbags.append(AIRBAGS.text)
    except:
        airbags.append('-') 
    
    try:
        POWERSTERING= driver.find_element_by_xpath("//div[contains(text(),'Power steering')]")
        Power_Stering.append(POWERSTERING.text)
    except:
        Power_Stering.append('-') 
    
    
    try:
        LOCKSYSTEM= driver.find_element_by_xpath("//div[contains(text(),'Lock System')]")
        Lock.append(LOCKSYSTEM.text)
    except:
        Lock.append('-') 
        
    try:
        PRIZE= driver.find_element_by_xpath('//div[@class="_3FkyT"]')
        prize.append(PRIZE.text)
    except:
        prize.append('-') 
        
```

```python
## scolled down manually to get more data

product_urls_carkho=[]
for i in driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]//a'):
    product_urls_carkho.append(i.get_attribute("href"))  
```

```python
def find_duplicates(sequence):
  first_seen = set()
  first_seen_add = first_seen.add  
  duplicates = set(i for i in sequence if i in first_seen or first_seen_add(i) )
  # turn the set into a list (as requested)
  return duplicates 


print(find_duplicates(product_urls_carkho))

#This shows that There are no duplicates
```

```python
#Make empty lists
General_Features = []   
General_Features_2 = []  
Spaecial_Features = []
Spaecial_Features_2=[]
prize=[]
prize_2=[]


for url in product_urls_carkho:
    driver.get(url)
    time.sleep(2)
    
    # Extracting Brand from xpath
    try:
        GF = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div')    
        General_Features.append(GF.text)
    except:
        General_Features.append('-')    
    time.sleep(2)
   
    try:
        GF_2 = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div')    
        General_Features_2.append(GF_2.text)
    except:
        General_Features_2.append('-')    
    time.sleep(2)
        
    
    try:
        SF = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[4]/div/div/div/div/div[1]/div/div/div[2]/div[1]/div/section/div/div')    
        Spaecial_Features.append(SF.text)
    except:
        Spaecial_Features.append('-')    
    time.sleep(2)
    
    try:
        SF_2 = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div/div/div[1]/div[6]/div/div/div')    
        Spaecial_Features_2.append(SF_2.text)
    except:
        Spaecial_Features_2.append('-')    
    time.sleep(2)
    
    
    try:
        Prize= driver.find_element_by_xpath('//span[@class="amount"]')    
        prize.append(Prize.text)
    except:
        prize.append('-')    
    time.sleep(2)
    
    try:
        Prize_2= driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div/div/div[2]/div[1]/div/div/div/div[1]/div[2]/div/div[1]')    
        prize_2.append(Prize_2.text)
    except:
        prize_2.append('-')    
    time.sleep(2)
    
    
```

```python
df1=pd.DataFrame({})
df1['General_Features_2']=General_Features_2
df1['Spaecial_Features_2']=Spaecial_Features_2
df1['General_Features']=General_Features
df1['Spaecial_Features']=Spaecial_Features
df1['prize']=prize
df1['prize_2']=prize_2
df1
```

```python
writer=pd.ExcelWriter('Ahmedabad_cardekho_CAR_DATA.xlsx')
df1.to_excel(writer,"Amedabad_cardekho_CAR_DATA", header=True, index= False)
writer.save()
```
