```python
driver=webdriver.Chrome(r'C:\Users\surendar.reddy\Documents\data Science projet_Data traine\chromedriver.exe')
time.sleep(3)
```

```python
url = "https://www.flipkart.com/"
driver.get(url)
```

```python
names= "Laptops,Phones,Headphones,smart watches,Cameras,Printers,Monitors,Home theater,Router"

res =names.split(',') 
pdt_urls=[]

for i in res:
    search_ITEM=driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
    search_ITEM.send_keys(i)
    search_ITEM=driver.find_element_by_xpath('//button[@class="L0Z3Pu"]').click()
    time.sleep(2)
    
    try:
        for urls in driver.find_elements_by_xpath('//a[@class="_1fQZEK"]'):
            pdt_urls.append(urls.get_attribute("href"))   
    except:
        for urls in driver.find_elements_by_xpath('//a[@class="_2rpwqI"]'):
                pdt_urls.append(urls.get_attribute("href")) 
                
    try:
        for urls in driver.find_elements_by_xpath('//a[@class="s1Q9rs"]'):
            pdt_urls.append(urls.get_attribute("href"))  
    except:
        pass
                 
    clear_ITEM1=driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input').clear()
```

```python
len(pdt_urls)
```

```python
pdt_urls
```

```python

review=[]
comment=[]
for u in pdt_urls:
    driver.get(u)
    time.sleep(1)
    
    try:
        time.sleep(1)
        morereviews=driver.find_element_by_xpath('//button[@class="_2KpZ6l _1FH0tX"]').click()
        time.sleep(1)
    except:
         print("No comments")
        
    for x in driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]'):
        try:
            review.append(x.text)
        except:
            review.append('-')
    
    for a in driver.find_elements_by_xpath('//p[@class="_2-N8zT"]'):
        try:                                
            comment.append(a.text)
        except:
            comment.append('-')
```
